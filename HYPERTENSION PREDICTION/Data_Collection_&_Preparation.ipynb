#Categorical Data Encoding

```python
nominal_features = ['Gender', 'History', 'Patient', 'TakeMedication',
                    'BreathShortness', 'VisualChanges', 'NoseBleeding', 'ControlledDiet']

ordinal_features = [f for f in df.columns if f not in nominal_features]
ordinal_features.remove('Stages')

print(nominal_features)
print(ordinal_features)
```

```python
for col in nominal_features:
    if set(df[col].unique()) == set(['Yes', 'No']):
        df[col] = df[col].map({'No': 0, 'Yes': 1})
    elif col == 'Gender':
        df[col] = df[col].map({'Male': 0, 'Female': 1})


df['Age'] = df['Age'].map({'18-34': 1, '35-50': 2, '51-64': 3, '65+': 4})

df['Severity'] = df['Severity'].replace({'Mild': 0, 'Moderate': 1, 'Severe': 2, 'Sever': 2})

df['Whendiagnoused'] = df['Whendiagnoused'].map({
    '<1 Year': 1,
    '1-5 Years': 2,
    '>5 Years': 3
})

df['Stages'] = df['Stages'].map({
    'NORMAL': 0,
    'HYPERTENSION (Stage-1)': 1,
    'HYPERTENSION (Stage-2)': 2,
    'HYPERTENSIVE CRISIS': 3
})

df['Systolic'] = df['Systolic'].map({
    '100 - 110': 0,
    '111 - 120': 1,
    '121 - 130': 2,
    '130+': 3})

df['Diastolic'] = df['Diastolic'].map({
    '70 - 80': 0,
    '81 - 90': 1,
    '91 - 100': 2,
    '100+': 3
})
```
